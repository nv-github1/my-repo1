```python
import csv
from datetime import datetime

# a var to store the user's monthly budget and a list to hold all expense as dictionaries

expenses = []
monthly_budget = 0.0

# we will check formatting and catch any error

def add_expense():
    while True:
        date = input("Enter the date (YYYY-MM-DD): ")
        try:
            datetime.strptime(date, "%Y-%m-%d")
            break
        except ValueError:
            print(" Invalid date format. Please try again.")

    category = input("Enter the category - can be Food, Travel, etc: ")

    while True:
        try:
            amount = float(input("Enter the amount: "))
            break
        except ValueError:
            print(" Invalid input. Please enter a number.")

    description = input("Enter a description: ")

    # Creating an expense dictionary with this
    expense = {
        "date": date,
        "category": category,
        "amount": amount,
        "description": description
    }

    expenses.append(expense)
    print(" Expense added successfully!\n")

# expenses added to dictionary till now.
# next part - viewing my expenses

def view_expenses():
    if not expenses:
        print("No expenses found as yet.\n")
        return

    print("\nThe expenses till now are:")

    count = 1
    for exp in expenses:
        print("Expense " + str(count) + ":")
        print("  Date: " + exp["date"])
        print("  Category: " + exp["category"])
        print("  Amount: " + str(exp["amount"]))
        print("  Description: " + exp["description"])
        print("------------------------")
        count = count + 1

    print()

# setting my monthly budget

def set_budget():
    global monthly_budget
    try:
        monthly_budget = float(input("Enter your monthly budget: "))
        print(f" your monthly budget is {monthly_budget}\n")
    except ValueError:
        print(" Invalid input. Please enter a number.\n")


# keeping a track of the budget and comparing my expenses with it

def track_budget():
    if monthly_budget == 0:
        print(" Please set a budget first before you shop.\n")
        return

    total = sum(exp['amount'] for exp in expenses)
    print(f" Total expenses so far: {total}")
    if total > monthly_budget:
        print(" You have exceeded your budget!\n")
    else:
        print(f" You have {monthly_budget - total} left for the month.\n")


# Saving my expenses to a CSV file - preparing by putting header first, then writing the rows

def save_expenses():



    rows = [["Date", "Category", "Amount", "Description"]]
    for exp in expenses:
        rows.append([exp['date'], exp['category'], exp['amount'], exp['description']])

    with open("expenses.csv", "w", newline="") as file:
        for row in rows:

  # convert all to string and join with commas - tried this using Google and Chatgpt help

            file.write(",".join(map(str, row)) + "\n")

    print("The expenses are saved to expenses.csv\n")

# we are loading our expenses, also 1st row is the header, so skipping it

def load_expenses():
    try:
        with open("expenses.csv", "r") as file:
            lines = file.readlines()
            for line in lines[1:]:
                date, category, amount, description = line.strip().split(",")
                expenses.append({
                    "date": date,
                    "category": category,
                    "amount": float(amount),
                    "description": description
                })
        print(" previous expenses are loaded now!\n")
    except FileNotFoundError:
        print("No saved expenses found. Starting a fresh file or tracker now.\n")

# creating an interactive menu

def my_menu():
    load_expenses()

    while True:
        print("     Expense Tracker Menu      ")
        print("1. Add Expense")
        print("2. View Expense")
        print("3. Set or Track Budget")
        print("4. Save Expense")
        print("5. Exit")
        choice = input("Choose an option between 1 and 5: ")

        if choice == "1":
            add_expense()
        elif choice == "2":
            view_expenses()
        elif choice == "3":
            if monthly_budget == 0:
                set_budget()
            track_budget()
        elif choice == "4":
            save_expenses()
        elif choice == "5":
            save_expenses()
            print("Yay! The expenses have been saved!")
            break
        else:
            print(" Not a valid choice. Please try again.\n")


# running the program here itself

my_menu()


```
